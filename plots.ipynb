```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
file_name = 'density_results.dat' 
all_data = np.loadtxt(file_name,skiprows=1) 
all_data = all_data.transpose() 
x = all_data[0] 
density_0 = all_data[1] 
density_1 = all_data[2] 
density_2 = all_data[3] 
density_3 = all_data[4] 
```

```python
plt.plot(x, density_0, label = "t = 0")
plt.plot(x, density_1, label = "1/3 the time") 
plt.plot(x, density_2, label = "2/3 the time")
plt.plot(x, density_3, label = "final time")
plt.xlabel('x')
plt.ylabel('density') 
plt.legend()
plt.show()
```

```python
file_name = 'time_results.dat' 
all_data = np.loadtxt(file_name,skiprows=1) 
all_data = all_data.transpose() 
t = all_data[0] 
position = all_data[1] 
position_squared = all_data[2] 
standard_deviation = all_data[3] 
analytic_deviation = all_data[4] 
```

```python
plt.plot(t, position, label = "expectation value <x>") 
plt.xlabel('t')
plt.ylabel('<x>') 
plt.legend()
plt.show()
```

```python
plt.plot(t, position_squared, label = "expectation value <x²>") 
plt.xlabel('t')
plt.ylabel('<x²>') 
plt.legend()
plt.show()
```

```python
plt.plot(t, standard_deviation, label = "numerical result") 
plt.plot(t, analytic_deviation, label = "analytic result") 
plt.xlabel('t')
plt.ylabel('σ') 
plt.legend()
plt.show()
```
